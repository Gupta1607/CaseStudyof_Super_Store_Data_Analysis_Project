```python
def Viz_Bar(df1 , cx , cy ,t='',rot="horizontal" , x=True ):
    if(x is False):
        plt.rcdefaults()
        fig, ax = plt.subplots()
        ax.barh(df1.head(10)[cx], df1.head(10)[cy], align='center', color=['#F05454','#30475E' ,'#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'] )
        ax.set_yticks(df1.head(10)[cx], labels=df1.head(10)[cx])
        ax.invert_yaxis()  # labels read top-to-bottom
        ax.set_xlabel(cy)
        ax.set_title(t)
        plt.xticks(rotation = rot)
        plt.show()
    else :
        plt.bar(df1.head(10)[cx] ,df1.head(10)[cy] ,  color=['#F05454','#30475E' ,'#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf'] )
        plt.xlabel(cx)
        plt.ylabel(cy)
        plt.title(t)
        plt.xticks(rotation = rot)
        plt.show()
```

```python
print(sales_data['Order Date'][0])
print(sales_data['Order Date'][0].year , " " , sales_data['Order Date'][0].month ," " ,sales_data['Order Date'][0].day)
```

Monthly dataset of Sales and Profit

```python
monthly_data = sales_data.groupby(['Order Date'], as_index=False).sum()[['Order Date' , 'Sales' , 'Profit']]

# Set the Order Date column as the index of the dataframe
monthly_data = monthly_data.set_index('Order Date')

# Resample the data into monthly intervals
monthly_data = monthly_data.resample('M').sum()

monthly_data.reset_index(inplace=True)
# Transforming Column For Clarification
monthly_data["Order Date"] = [str(k.month)+'/'+str(k.year) for k in monthly_data['Order Date'] ]
monthly_data.rename(columns = {"Order Date":'Month'}, inplace = True)
monthly_data = monthly_data.set_index('Month')


monthly_data
```

Monthly Sales From 1/2014 to 12/2017

```python
plt.figure(figsize=(25,8))
plt.plot(monthly_data['Sales'])
plt.axvline(monthly_data.sort_values(by = ["Sales"] , ascending = False).reset_index()['Month'][0], color='black')

plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.title("MONTHLY SALES")
plt.xticks(rotation='vertical')
plt.show()
```

Monthly Profit From 1/2014 to 12/2017

```python
# Plot
plt.figure(figsize=(25,8))
plt.plot(monthly_data['Profit'])
plt.axvline(monthly_data.sort_values(by = ["Profit"] , ascending = False).reset_index()['Month'][0], color='black')

plt.xlabel("Month/Year")
plt.ylabel("Sales")
plt.title("MONTHLY PROFIT")
plt.xticks(rotation='vertical')
plt.show()
```

Yearly Sales and Profit dataset

```python
yearly_data = sales_data.groupby(['Order Date'], as_index=False).sum()[['Order Date' , 'Sales' , 'Profit']]

# Set the Order Date column as the index of the dataframe
yearly_data = yearly_data.set_index('Order Date')

# Resample the data into yearly intervals
yearly_data = yearly_data.resample('Y').sum()

yearly_data.reset_index(inplace=True)
# Transforming Column For Clarification
yearly_data["Order Date"] = [str(k.year) for k in yearly_data['Order Date'] ]
yearly_data.rename(columns = {"Order Date":'Year'}, inplace = True)
yearly_data = yearly_data.set_index('Year')


yearly_data
```

Yearly Sales from 2014 to 2017

```python
# Plot
plt.figure(figsize=(25,8))
plt.plot(yearly_data['Sales'])
plt.xlabel("Year")
plt.ylabel("Sales")
plt.title("YEARLY SALES")
plt.show()
```

Yearly Profit from 2014 to 2017

```python
# Plot
plt.figure(figsize=(25,8))
plt.plot(yearly_data['Profit'])
plt.xlabel("Year")
plt.ylabel("Profit")
plt.title("YEARLY PROFIT")
plt.show()
```
